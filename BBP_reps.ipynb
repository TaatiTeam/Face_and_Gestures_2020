```python
import pandas as pd
import numpy as np
import os
```

```python
path_tables = r'C:\Users\GuarinD\Documents\GitHub\Face_and_Gestures_2020\data\audio'
Files = os.listdir(path_tables)            
ext=('.Table')
Files_Tables = [i for i in Files if i.endswith(tuple(ext))]
         
         
paths = [r'C:\Users\GuarinD\Documents\GitHub\Face_and_Gestures_2020\data\Healthy', r'C:\Users\GuarinD\Documents\GitHub\Face_and_Gestures_2020\data\Parkinsons']
#r'C:\Users\GuarinD\Documents\GitHub\Face_and_Gestures_2020\example'
Files_select = []
for k, path in enumerate(paths):
    Files = os.listdir(path)            
    ext=('.csv')
    Files = [i for i in Files if i.endswith(tuple(ext))]
    #Files_select = []
    for k,f in enumerate(Files):
        Files[k] = os.path.join(path,f)
        if ('frameInfoColor' in f) and ('BBP' in f): 
            Files_select.append(Files[k])


            
            
info_start_time = pd.read_csv(r"C:\Users\GuarinD\Documents\GitHub\Face_and_Gestures_2020\data\info_sync.csv", index_col=0 )
            
all_res = {}            
for aux, file in  enumerate(Files_select):
    datum = []
    _,f = os.path.split(file)
    
    datum.append(f[0:4])
    datum.append('BBP')
    f = f[:-19]
    
    
    table_info = pd.read_csv(os.path.join(path_tables, f+'.Table'))
    frames_info = pd.read_csv(file, index_col=0)
    
    info_about_that_file = info_start_time[info_start_time['File']==f].values
    
    st_time = frames_info.Frame_Time_Stamp[frames_info['Frame_Number_in_Video']==info_about_that_file[0][-1]-1].values[0]
    
    time_vector = frames_info.Frame_Time_Stamp.values[:]
    
    all_trials = table_info[['tmin','tmax']].values
    for trial in all_trials:
        st_trial = trial[0] + st_time
        end_trial = trial[1] + st_time
        
        time_st = time_vector[(np.abs(time_vector  - st_trial)).argmin()]
        time_end = time_vector[(np.abs(time_vector  - end_trial)).argmin()]
        
        frame_st =  frames_info.Frame_Number_in_Video[frames_info.Frame_Time_Stamp==time_st].values[0]
        frame_end =  frames_info.Frame_Number_in_Video[frames_info.Frame_Time_Stamp==time_end].values[0]
        
        datum.append(str(frame_st)+':'+str(frame_end))
        
    
    all_res[aux]=datum
    
    
    
```

```python
df = pd.DataFrame.from_dict(all_res, orient='index')

cols = df.columns
new_names = ['Subject_ID','Task']
for i in range(2):
    df.rename(columns={cols[i]:new_names[i]}, inplace = True)
df.to_csv(r"C:\Users\GuarinD\Documents\GitHub\Face_and_Gestures_2020\data\results_BBP_new.csv")
```
